```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.log(np.cosh(x))

def d_f(x):
    h = 1e-5
    return (f(x+h)-f(x-h))/(2*h)

def Q(x,y):
    return f(y) + np.multiply(d_f(y),x-y) + 0.5 * (x-y) ** 2

x = np.linspace(-5,5,100)
f_x = f(x)
q = Q(x,-2)
q_3 = Q(x,3)

plt.plot(x,f_x, label = 'f(x)')
plt.plot(x,q, label = 'Q(x,-2)')
plt.plot(x,q_3, label = 'Q(x,3)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

Auxillary Function

```python
iter = 20
xn = -2
xn_3 = 3
xn_list = np.array([-2])
xn_list_3 = np.array([3])
for n in range(1,iter):
    xn -= np.tanh(xn)
    xn_3 -= np.tanh(xn_3)
    xn_list = np.append(xn_list,xn)
    xn_list_3 = np.append(xn_list_3,xn_3)

plt.plot(range(iter), xn_list, 'o-', label = 'x0 = -2')
plt.plot(range(iter), xn_list_3, 'x-', label = 'x0 = 3')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

Newton's Method

```python
xn_list = np.array([-2])
xn_list_3 = np.array([3])

for n in range(1,iter):
    xn -= np.sinh(xn) * np.cosh(xn)
    xn_3 -= np.sinh(xn_3) * np.cosh(xn_3)
    xn_list = np.append(xn_list,xn)
    xn_list_3 = np.append(xn_list_3,xn_3)

plt.plot(range(iter), xn_list, 'o-', label = 'x0 = -2')
plt.plot(range(iter), xn_list_3, 'x-', label = 'x0 = 3')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
```

```python
def Newtons_method(xn):
    return xn-np.sinh(xn)*np.cosh(xn)

def Newtons_method_update(x0, n):
    xn_l = [x0]
    for i in range(len(n)-1):
        xn_l.append(Newtons_method(xn_l[-1]))

    return xn_l

n = np.arange(0, 20)
x0 = -2
x0_3 = 3
xn_2_list = Newtons_method_update(x0, n)
xn_3_list = Newtons_method_update(x0_3, n)

plt.figure()
plt.title("Update rule of Newton's method, with x0=-2")
plt.plot(n, xn_2_list, 'o-')
plt.ylabel('Xn')
plt.xlabel('n')

plt.figure()
plt.title("Update rule of Newton's method, with x0=3")
plt.plot(n, xn_3_list, 'o-')
plt.ylabel('Xn')
plt.xlabel('n')
```

```python
def updateRuleNewton(xn):
    return xn-np.sinh(xn)*np.cosh(xn)

def getUpdatedXnByNetwon(x0, n):
    arr_xn = [x0]
    for i in range(len(n)-1):
        arr_xn.append(updateRuleNewton(arr_xn[-1]))

    return arr_xn

# non-converging example
n = np.arange(0, 11)
x0_2 = -3
arr_xn_2 = getUpdatedXnByNetwon(x0_2, n)

plt.plot(n, arr_xn_2)
plt.ylabel('Xn')
plt.xlabel('n')
plt.title("Update rule of Newton's method, with x0=-3")
```
