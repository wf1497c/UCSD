```python
import numpy as np
from sklearn.preprocessing import normalize
import random
import matplotlib.pyplot as plt
```

```python
def define_A(M,N):
    A = np.random.normal(0,1,[M,N])
    return normalize(A, axis = 0) 

def define_x(s,N):
    i = random.sample(range(N),s)
    x = np.zeros(N)
    for idx in i:
        r = np.random.rand()
        if r >0.5: #randomly drawn from [-10,-1] || [1,10]
            x[idx]=(r-0.5)*18+1 
        else:
            x[idx]=(r-0.5)*18-1
    return x.T, i 
```

```python
def OMP(A, y, N, thresh=0.001, stop=np.infty):
    resi = y
    x_pred = np.zeros(N)
    lambda_list = []
    i = 0
    iteration = 0
    # Control stop interation with norm thresh or sparsity
    while np.linalg.norm(resi) > thresh and iteration < stop:
        scores = A.T.dot(resi) # score of each atoms
        lambda_list.append(np.argmax(abs(scores))) # atom with highest score
        basis = A[:, lambda_list]

        x_pred[lambda_list] = np.linalg.pinv(np.dot(basis.T,basis)).dot(basis.T).dot(y) # Least square solution for y=Ax
        
        # Compute the residual
        resi = y - A.dot(x_pred)
        iteration += 1
    return x_pred.T, lambda_list

def Error_normalized(x,x_pred):
    return np.linalg.norm(x-x_pred) / np.linalg.norm(x)
```

```python
def plot_phase_transition(x, y, z, title):
    plt.figure()
    X, Y = np.meshgrid(x, y)
    im = plt.pcolormesh(X, Y, z.T, vmin=np.min(z), vmax=np.max(z), shading='auto')
    plt.colorbar(im)
    plt.xticks(x[:])
    plt.yticks(y[:])
    plt.xlabel('s_max')
    plt.ylabel('M')
    plt.title(title)
```

Normalized error and probability of Exact Support Recovery in OMP, as results shown below, we can observe a sharp transition region where the prob. of ESR quickly transitions from a large value(close to 1) to a very small value(close to 0). When s_max and M continually grows, the transition region becomes blurred. We can also see that as M increases, which means our measurement rate increases with fixed value N, we tend to achieves higher prob. of ESR under the same sparsity rate. On the contrary, as s_max increases, which means the sparsity grows with fixed value N, lower prob. of ESR has been shown. Because in compressed sensing, the more information we get, we are more likely to recover the signal.

```python
M = range(20,100,5)
N = [20,50,100]
times = 100

def noiseless(N):
    s_max_range = range(N // 2, N, 3)
    map_n = np.zeros((len(s_max_range),len(M)))
    for m_i in range(len(M)):
        m = M[m_i]
        for s_max_i in range(len(s_max_range)):
            s_max = s_max_range[s_max_i]
            e = 0
            for i in range(times):
                s = np.random.randint(low=1,high = s_max)
                A = define_A(m,N)
                x, ind = define_x(s,N)
                y = A.dot(x)
                x_pred, lamda = OMP(A, y, N)
                e += Error_normalized(x,x_pred)
            e /= times # average of 2000 results
            map_n[s_max_i, m_i] = e
    
    return s_max_range, map_n

titles = ["Normalized Error: N = 20", "Normalized Error: N = 50", "Normalized Error: N = 100"]

for n_i in range(len(N)):
    s_max, tmp = noiseless(N[n_i])
    plot_phase_transition(s_max, M, tmp, titles[n_i])
```

In noisy case, I chose two values of sigma in Gaussian: 0.0001 and 0.01. As results shown, the existence of noise affects the performance of recovery. As noise levels increases with higher varaince, the boundary in transition region becomes more blurred. This is intuitive since less variance noise is less effective to signal recovery.

```python
def noisy_exp(N, sigma):
    s_max_range = range(N // 2, N, 3)
    success_map = np.zeros((len(s_max_range),len(M)))
    for m_i in range(len(M)):
        m = M[m_i]
        A = define_A(m,N)
        for s_max_i in range(len(s_max_range)):
            s_max = s_max_range[s_max_i]
            t = 0
            for i in range(times):
                s = np.random.randint(low=1,high = s_max)
                x, ind = define_x(s,N)
                y = A.dot(x) + np.random.normal(0, sigma, m) # noisy case
                x_pred, lamda = OMP(A, y, N, stop = s)
                if Error_normalized(x, x_pred) < 0.001:
                    t += 1
            t /= times
            success_map[s_max_i, m_i] = t
    
    return s_max_range, success_map

sigma = [0.0001, 0.01]
titles = [["Probability of Successful Recovery(sigma = 0.0001): N = 20", "Probability of Successful Recovery(sigma = 0.0001): N = 50", "Probability of Successful Recovery(sigma = 0.0001): N = 100"], ["Probability of Successful Recovery(sigma = 0.01): N = 20", "Probability of Successful Recovery(sigma = 0.01): N = 50", "Probability of Successful Recovery(sigma = 0.01): N = 100"]]

for n_i in range(len(N)):
    for s_i in range(2):
        sig = sigma[s_i]
        s_max, tmp = noisy_exp(N[n_i], sig)
        plot_phase_transition(s_max, M, tmp, titles[s_i][n_i])
```

Decode a Compressed Message

```python
import scipy.io

data = scipy.io.loadmat('hw3_data.mat')
A1 = data['A1'] # M = 960, N = 14400
A2 = data['A2']
A3 = data['A3']
y1 = data['y1'][:,0] # M = 960
y2 = data['y2'][:,0]
y3 = data['y3'][:,0]
```

In part (a), I think this compressed image is a periodic wave with noises since I can clearly see a "central band" in the signal(960*1). Because Y = <A,X>, which means compressed image Y is transformation of X, I guess message in X is also a periodic wave. (But Im not sure what it means)

```python
plt.figure()
plt.plot(y1)
plt.title("Y1")
plt.figure()
plt.plot(y2)
plt.title("Y2")
plt.figure()
plt.plot(y3)
plt.title("Y3")
```

In part (b), I reconstruct X with OMP. Before reshaping, figures show that 1-D form of X is similar to periodic square wave.

```python
x1, lambda1 = OMP(A1, y1, A1[1].shape)
x2, lambda2 = OMP(A2, y2, A2[1].shape)
x3, lambda3 = OMP(A3, y3, A3[1].shape)

f = plt.figure(figsize=(15,5))
f.add_subplot(131)
plt.title('OMP: X1')
plt.plot(x1)
f.add_subplot(132)
plt.title('OMP: X1')
plt.plot(x2)
f.add_subplot(133)
plt.title('OMP: X1')
plt.plot(x3)
```

To check my reconstruction, inner product of Ai and xi is checked in below as reconstruction error, and all such error approach zero.

```python
f = plt.figure(figsize=(15,5))
f.add_subplot(131)
plt.title('Reconstruction error by OMP: X1')
plt.plot(A1.dot(x1) - y1)
f.add_subplot(132)
plt.title('Reconstruction error by OMP: X2')
plt.plot(A2.dot(x2) - y2)
f.add_subplot(133)
plt.title('Reconstruction error by OMP: X3')
plt.plot(A3.dot(x3) - y3)
```

For least square solution, the error is similar to OMP result, which indicates two methods are appropriate to this question.

```python
x1_LS = np.linalg.pinv(A1).dot(y1)
x2_LS = np.linalg.pinv(A2).dot(y2)
x3_LS = np.linalg.pinv(A3).dot(y3)

f = plt.figure(figsize=(15,5))
f.add_subplot(131)
plt.title('Reconstruction error by Least Square Sol.: X1')
plt.plot(A1.dot(x1_LS) - y1)
f.add_subplot(132)
plt.title('Reconstruction error by Least Square Sol.: X2')
plt.plot(A2.dot(x2_LS) - y2)
f.add_subplot(133)
plt.title('Reconstruction error by Least Square Sol.: X3')
plt.plot(A3.dot(x3_LS) - y3)
```

In part (c), I obsereve that corrupted image Y3 gives best result. I think the reason is dimsension of Y3 is much larger than Y1 and Y2. This dimension is N in previous experiment: larger N gives higher probability of sucessful reconstruction. 

```python
f = plt.figure(figsize=(20,15))
f.add_subplot(131)
plt.title('2-D Reconstruction by OMP: X1')
plt.imshow(np.reshape(x1, [90,160]), cmap = 'gray')
f.add_subplot(132)
plt.title('2-D Reconstruction by OMP: X2')
plt.imshow(np.reshape(x2, [90,160]), cmap = 'gray')
f.add_subplot(133)
plt.title('2-D Reconstruction by OMP: X3')
plt.imshow(np.reshape(x3, [90,160]), cmap = 'gray')
```
